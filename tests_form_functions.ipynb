```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def m1(w,w_hat):
    #return np.sqrt(np.real(w.T*w_hat)^2+np.imag(w.T*w_hat)^2)
    return np.linalg.norm(w.T@w_hat)

def m2(w,w_hat):
    prod = np.conj(w).T@w_hat
    return np.linalg.norm(prod)
```

```python
def rand_complex_vector(n=1):
    """
    returns a complex vector of dimension n x 1 (defaults to 1 x 1), complex elements have norm 1
    """
    ret = np.empty(n,dtype=np.complex_)
    for i in range(n):
        ang = 2*np.pi*random.random()
        ret[i] = np.exp((0+1j)*ang)
    return ret
```

```python
nb_sam = 50
samples_m1 = np.empty(nb_sam)
samples_m2 = np.empty(nb_sam)
for i in range(nb_sam):
    w = rand_complex_vector(10)
    w_hat = rand_complex_vector(10)
    samples_m1[i] = m1(w,w_hat)
    samples_m2[i] = m2(w,w_hat)

plt.subplot(1,2,1)
plt.plot(range(nb_sam),samples_m1,color='b',label='m1')
plt.subplot(1,2,2)
plt.plot(range(nb_sam),samples_m2,color='r',label='m2')
plt.legend
```
